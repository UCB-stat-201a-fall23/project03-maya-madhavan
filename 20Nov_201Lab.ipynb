Simulate one single realization of the chain, that is, starting from X0 = 1, update the value of Xi using the probabilities defined by the process.

```python
transition_matrix = np.array([
    [0.2, 0.7, 0.1],
    [0.2, 0.5, 0.3],
    [0.2, 0.4, 0.4]
])
```

```python
current_state = 1
```

# 2b
Starting now from an initial probability distribution π0 on the nodes, compute the value of πTi = πT0 * Pi the probability distribution at time i. Show that πi → π∞ and make plot of i vs ∥πi − π∞∥2. Generate this plot for at least two different initial conditions π0 and compare

```python
pi_inf = np.array([9/45, 23/45, 13/45])
pi_0 = np.array([1/3, 1/3, 1/3])
```

```python
i_axis = []
euclid_distance = []
pi_0_transpose = np.transpose(pi_0)

for i in range(1, 30):
    i_axis.append(i)
    transition_matrix_i = np.linalg.matrix_power(transition_matrix, i)
    pi_t = np.dot(pi_0, transition_matrix_i)  # Using pi_0 directly
    distance = math.dist(pi_t, pi_inf)
    euclid_distance.append(distance)
```

```python
plt.plot(i_axis, euclid_distance)
plt.yscale("log")
```

# 3a
Starting from each one of X0 = 1 and X0 = 2, run multiple simulation of the Markov chain (Problem 1, part b) until Xi = 3 and store the arrival time until this happens. Make a histogram of the arrival time for both X0 = 1 and X0 = 2 and compute the mean

```python
arrival_time = []

for i in range(1, 100000):
    current_state = 1
    states = []
    while current_state != 3:
        current_state = np.random.choice([1, 2, 3], p=transition_matrix[current_state-1])
        states.append(current_state)
    arrival_time.append(len(states))
```

```python
plt.hist(arrival_time, bins=np.arange(0.5, 30, 1));
```

```python
np.mean(arrival_time)
```
